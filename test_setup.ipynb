```python
import fastf1
import pandas as pd

# Create a folder for caching data so it doesn't download every time
fastf1.Cache.enable_cache('cache') 

# Load the qualifying session for the 2023 Bahrain Grand Prix
session = fastf1.get_session(2023, 'Bahrain', 'Q')
session.load()

# Get the fastest lap data
fastest_lap = session.laps.pick_fastest()

print(f"Fastest Lap in Bahrain 2023: {fastest_lap['LapTime']} by {fastest_lap['Driver']}")
```
